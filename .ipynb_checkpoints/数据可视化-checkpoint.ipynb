## Example 


`````python
x=np.linspace(0,10,1000) # 0到10分成1000份
y=np.sin(x)
z=np.cos(x**2)
`````

`````python
plt.figure(num=1,figsize=(8,4))
plt.plot(x,y,label="$sin(x)",color="red",linewidth=2)#图形关系变量，标注，线色（默认蓝色），线宽
plt.plot(x,z,"g--",label="cos(x^2)")#  线形：  -，细线，--，虚线，+,＋号，o，O形，^,^形
plt.xlabel("Times(s)") #X轴标题
plt.ylabel("Volt") 
plt.title("Matplotlib Example") #图形标题
plt.ylim(-1.2,1.2) # Y轴值上下限
plt.legend()   #   图示,默认值为plt.plot中的label，可按顺序指定,如plt.legend(['sin(x)','cos(x^2)'])
plt.grid(True)  # 显示网格
fig.patch.set_alpha(0.8) #设置图片的透明度
plt.show() #激活图像
`````

 线型：
 - -，细线
 - --，虚线
 - +,＋号
 - o，O型
 - ^,^型

`````python
def f(t):    
    return np.exp(-t)*np.cos(2*np.pi*t)    

t1 = np.arange(0.0,5.0,0.1) 
t2 = np.arange(0.0,5.0,0.02)  
`````

`````python
plt.plot(t1,f(t1),'bo') 

plt.show()
`````

`````python

plt.plot(t1,f(t1),'bo',t2,f(t2),'k') 
plt.show()

`````

`````python
plt.plot(t2,np.cos(2*np.pi*t2),'g--')
plt.show()
`````

`````python

plt.subplot(221,axisbg='g') 

# subplot(nrows, cols, plot_number)
# 2x2共4子图，把此图放在第一个

plt.plot(t1,f(t1),'bo',t2,f(t2),'k') 

plt.subplot(224,axisbg='y') 

# 2x2共4子图，把此图放在第四个

plt.plot(t2,np.cos(2*np.pi*t2),'g--') 

plt.show()
`````

## 子图

`````python
cols = ['Column {}'.format(col) for col in range(1, 4)]
rows = ['Row {}'.format(row) for row in ['A', 'B', 'C']]

fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(6, 6))

for ax, col in zip(axes[0], cols):
    ax.set_title(col)

for ax, row in zip(axes[:,0], rows):
    ax.set_ylabel(row, rotation=0, size='large')

fig.tight_layout()
plt.show()
`````

## 散点图

`````python
from pylab import *

n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

scatter(X,Y)
show()
`````

## 柱状图

`````python
from pylab import *

n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

bar(X, +Y1, facecolor='#9999ff', edgecolor='red')
bar(X, -Y2, facecolor='#ff9999', edgecolor='green')

for x,y in zip(X,Y1):
    text(x+0.4, y+0.05, '%.2f' % y, ha='center', va= 'bottom')

ylim(-1.25,+1.25)
show()
`````

## 等高线

`````python
from pylab import *

def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='jet')
C = contour(X, Y, f(X,Y), 8, colors='black', linewidth=.5)
show()
`````

## 灰度图

`````python
from pylab import *

def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 10
x = np.linspace(-3,1,3*n)
y = np.linspace(-3,3,3*n)

X,Y = np.meshgrid(x,y)
imshow(f(X,Y)), show()

show()
`````

## 极轴图

`````python
from pylab import *

axes([0,0,1,1])

N = 20
theta = np.arange(0.0, 2*np.pi, 2*np.pi/N)
radii = 10*np.random.rand(N)
width = np.pi/4*np.random.rand(N)
bars = bar(theta, radii, width=width, bottom=0.0)

for r,bar in zip(radii, bars):
    bar.set_facecolor( cm.jet(r/10.))
    bar.set_alpha(0.5)

show()
`````

## 网格

`````python
from pylab import *

axes = gca()
axes.set_xlim(0,4)
axes.set_ylim(0,3)
axes.set_xticklabels([])
axes.set_yticklabels([])

show()
`````

# 3D

`````python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')

show()
`````

## 其他

区间刻度的设置：
```
1.选择是否显示刻度值：x轴上，1为下，2为上；y轴上，1为左，2为右； 
for tick in ax.xaxis.get_major_ticks():   
    tick.label1On = True     
    tick.label2On = False  
for tick in ax.yaxis.get_major_ticks():   
    tick.label1On = False  
    tick.label2On = False
    
2.选择如何显示刻度 
ax.xaxis.set_ticks_position(‘none’) ax.yaxis.set_ticks_position(‘right’)

3.完全自定义坐标轴刻度及刻度值  
ax.set_xticks([-250, -200, -150, -100, -50, 0, 50, 100, 150, 200]) 
ax.set_xticklabels(['-250', '', '-150', '', '-50', '', '50', '', '150', '']) 
ax.set_yticks([-40, -20, 0, 20, 40, 60, 80]) ax.set_yticklabels(['', '', '', '', '', '', ''])  
        
3.分别设置x轴和y轴上刻度值的字体大小 
for tick in ax.xaxis.get_major_ticks():   
    tick.label1.set_fontsize(18) 
for tick in ax.yaxis.get_major_ticks():    
     tick.label1.set_fontsize(18) 
                
4.这也可以设置刻度值字体大小
matplotlib.rc(‘xtick’, labelsize=20) matplotlib.rc(‘ytick’, labelsize=20) 
````

# 其他图像处理库

  Pandas，做简单的绘图非常的灵巧，但是你需要学习Matplotlib来进行个性化设置。

  Seaborn，可以支持一些更复杂的可视化方案，但是依旧需要Matplotlib的知识来调整。颜色模型是非常好的加分。

  ggplot，期望很高，但是还需要经历“成长的痛苦”。

  Bokeh，是一个鲁棒的工具如果你想要设置你自己的可视化Web服务器，但是对于简单的场景可能有些功能过度。

  Pygal，是唯独可以产生可交互试svg图像和PNG文件的。然而它不如基于Matplotlib的实现的库更灵活。

  Plot.ly，能够产生最具交互性的图表。你可以在线下存储他们或者创建非常丰富的基于web的可视化图形。
